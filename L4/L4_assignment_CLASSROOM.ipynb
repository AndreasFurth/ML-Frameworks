## Task 1: Dataset and split
Load a dataset and create a reproducible train/test split.

```python
# TODO: Load a dataset (iris, breast_cancer, or diabetes)
from sklearn.datasets import load_iris, load_diabetes

# I produktion: 
# data = pd.load_csv("filnamn.csv)
# Den datan kan dock behöva mer förarbete
# Med färdigbyggda dataset, behövs sällan preprocessing
# MEN, EDA är fortfarande väldigt viktigt! 
# Om man har bra ETL / data engineers, kan datan vara användbar från början

data = load_iris(as_frame=True)

X = data.data
y = data.target
target_names = data.target_names

print("X: ", X)
print("y: ", y)
```

```python
X
```

```python
# TODO: Create train/test split with a fixed random_state

# Here we often use a function (train_test_split)
# The function splits our data neatly into piles of train and test
# We could do this manually

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# En sak att hålla 

print("X_train shape:", X_train.shape, "X_test shape:", X_test.shape)
```

```python
# Det är bra att dubbelkolla att man har alla klasser
# från hela datasetet kvar efter att man har splittat datan
# Man kan ha otur med splitten

y_train.value_counts()
```
